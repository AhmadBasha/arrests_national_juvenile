```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("arrests_national_juvenile.csv")

# we just make sure that we loaded the right data 
df.head()
```

```python
#Show the the count of rows and columns
df.shape
```

```python
# as we notice here we have many Nan in state_abb , so we need to check the missing values for all columns. 
df.describe()
```

```python
# to see the missing values for each column  , 
#As we can see here in "state_abbr "690 which is the same number of rows. That means there is no value for this column

df.isna().sum()
```

```python
# Delete "state_abbr"
del df["state_abbr"]
df
```

```python
# Other missing values , Because it is a period of time so fill forweads will be a great solution.
df["m_0_9"] = df["m_0_9"].fillna(method="ffill")
df["m_10_12"] = df["m_10_12"].fillna(method="ffill")
df["f_0_9"] = df["f_0_9"].fillna(method="ffill")
df["f_10_12"] = df["f_10_12"].fillna(method="ffill")
df["f_13_14"] = df["f_13_14"].fillna(method="ffill")
df["f_15"] = df["f_15"].fillna(method="ffill")
df["f_16"] = df["f_16"].fillna(method="ffill")
df["asian_pacific_islander"] = df["asian_pacific_islander"].fillna(method="ffill")
df["american_indian"] = df["american_indian"].fillna(method="ffill")
```

```python
#make a copy of the data after finishing the data preperation.
df_copy = df.copy()
```

```python
df_total = df_copy.copy()
df_total["total"] = df_total["total_male"] + df_total["total_female"]
df_total
```

```python
# group by offense_name taking the total and each range to compare later and we order them to take top 5
df_top7_offense = df_total.groupby('offense_name').agg({'total':'sum', \
                                                         'm_0_9':'sum', 'm_10_12':'sum', \
                                                        'm_13_14':'sum' , 'm_15':'sum', \
                                                        'm_16':'sum' , 'm_17':'sum', \
                                                         'f_0_9':'sum', 'f_10_12':'sum', \
                                                         'f_13_14':'sum' , 'f_15':'sum', \
                                                         'f_16':'sum' , 'f_17':'sum' \
                                                        })

df_top7_offense = df_top7_offense.sort_values(by='total', ascending=False).head(7).reset_index()
df_top7_offense
```

```python
# from the top 7 offenses we can see males having hiegher crimes than females for all age ranges except 
# Runaway from 13 to 16 , with 17 is still hiegher littile bit on male . 


# we will use it to know the limit for each age range and gender 
def spider_values(iloc_number_col, first_list=None ):
    values_2=first_list.iloc[:, iloc_number_col].tolist()
    values_2.append(values_2[0])
    return values_2

# for the graph limit 
values=df_top7_offense.iloc[:, 0].tolist()
values.append(values[0])

# styling 
plt.style.use("bmh")

#plot 
fig = plt.figure(figsize=(20,14))




#first plot 
ax=fig.add_subplot(321,polar=True)

#male value 
male_value = spider_values(first_list = df_top7_offense , iloc_number_col = 2)

ax.plot(values,male_value, color="blue", linewidth=1, linestyle='solid' , label="m_0_9")
ax.fill(values, male_value , color="blue" , alpha= 0.2)

#female value 
female_value = spider_values(first_list = df_top7_offense , iloc_number_col = 8)

ax.plot(values,female_value, color="red", linewidth=1, linestyle='solid', label="f_0_9")
ax.fill(values, female_value , color="red" , alpha= 0.2)

plt.yticks([5000, 10000 , 20000 , 40000, 60000], ["5000","10000", "20000", "40000", "60000"], 
           color="grey", size=8)
#plt.ylim(0,89999)

plt.legend(loc='lower left', bbox_to_anchor=(1, 1))

#---------------------------------------------

#second plot 
ax=fig.add_subplot(322,polar=True)

#male value 
male_value = spider_values(first_list = df_top7_offense , iloc_number_col = 3)

ax.plot(values,male_value, color="blue", linewidth=1, linestyle='solid' , label="m_10_12")
ax.fill(values, male_value , color="blue" , alpha= 0.2)

#female value 
female_value = spider_values(first_list = df_top7_offense , iloc_number_col = 9)

ax.plot(values,female_value, color="red", linewidth=1, linestyle='solid', label="f_10_12")
ax.fill(values, female_value , color="red" , alpha= 0.2)

plt.yticks([50000, 80000 , 100000 , 200000, 300000], ["50000","8000", "100000", "200000", "300000"], 
           color="grey", size=8)


plt.legend(loc='lower left', bbox_to_anchor=(1, 1))

#---------------------------------------------

#Third plot 
ax=fig.add_subplot(323,polar=True)

#male value 
male_value = spider_values(first_list = df_top7_offense , iloc_number_col = 4)

ax.plot(values,male_value, color="blue", linewidth=1, linestyle='solid' , label="m_13_14")
ax.fill(values, male_value , color="blue" , alpha= 0.2)

#female value 
female_value = spider_values(first_list = df_top7_offense , iloc_number_col = 10)

ax.plot(values,female_value, color="red", linewidth=1, linestyle='solid', label="f_13_14")
ax.fill(values, female_value , color="red" , alpha= 0.2)

plt.yticks([300000, 400000 , 600000 , 800000, 1000000], ["300000","400000", "600000", "800000", "1000000"], 
           color="grey", size=8)


plt.legend(loc='lower left', bbox_to_anchor=(1, 1))


#---------------------------------------------

#Fourth plot 
ax=fig.add_subplot(324,polar=True)

#male value 
male_value = spider_values(first_list = df_top7_offense , iloc_number_col = 5)

ax.plot(values,male_value, color="blue", linewidth=1, linestyle='solid' , label="m_15")
ax.fill(values, male_value , color="blue" , alpha= 0.2)

#female value 
female_value = spider_values(first_list = df_top7_offense , iloc_number_col = 11)

ax.plot(values,female_value, color="red", linewidth=1, linestyle='solid', label="f_15")
ax.fill(values, female_value , color="red" , alpha= 0.2)

plt.yticks([300000, 400000 , 600000 , 800000, 1000000], ["300000","400000", "600000", "800000", "1000000"], 
           color="grey", size=8)


plt.legend(loc='lower left', bbox_to_anchor=(1, 1))

#---------------------------------------------

#Fifth plot 
ax=fig.add_subplot(325,polar=True)

#male value 
male_value = spider_values(first_list = df_top7_offense , iloc_number_col = 6)

ax.plot(values,male_value, color="blue", linewidth=1, linestyle='solid' , label="m_16")
ax.fill(values, male_value , color="blue" , alpha= 0.2)

#female value 
female_value = spider_values(first_list = df_top7_offense , iloc_number_col = 12)

ax.plot(values,female_value, color="red", linewidth=1, linestyle='solid', label="f_16")
ax.fill(values, female_value , color="red" , alpha= 0.2)

plt.yticks([300000, 400000 , 600000 , 800000, 1000000 , 1200000], ["300000","400000", "600000", "800000", "1000000", "1200000"], 
           color="grey", size=8)


plt.legend(loc='lower left', bbox_to_anchor=(1, 1))

#---------------------------------------------

#sixth plot 
ax=fig.add_subplot(326,polar=True)

#male value 
male_value = spider_values(first_list = df_top7_offense , iloc_number_col = 7)

ax.plot(values,male_value, color="blue", linewidth=1, linestyle='solid' , label="m_17")
ax.fill(values, male_value , color="blue" , alpha= 0.2)

#female value 
female_value = spider_values(first_list = df_top7_offense , iloc_number_col = 13)

ax.plot(values,female_value, color="red", linewidth=1, linestyle='solid', label="f_17")
ax.fill(values, female_value , color="red" , alpha= 0.2)

plt.yticks([300000, 400000 , 600000 , 800000, 1000000 , 1200000 , 1400000], ["300000","400000", "600000", "800000", "1000000", "1200000" , "1400000"], 
           color="grey", size=8)


plt.legend(loc='lower left', bbox_to_anchor=(1, 1))



plt.tight_layout() 

fig.suptitle('Age range for both gander of top 6 crimes' , fontsize=18)
plt.show()

```

```python

#we group by offense_name and taking the sum of total_male , assign it to new var 
df_top10_offense_m = df_copy.groupby(["offense_name"]).total_male.sum()
# make it with descending order and take the top 10 
df_top10_offense_m = df_top10_offense_m.sort_values(ascending=False).head(10)
# here we reset the index to make it as dataframe 
df_top10_offense_m = df_top10_offense_m.reset_index()

df_top10_offense_m
```

```python
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html
# check the top 10 crimes for males. and check later with female.

fig = plt.figure(figsize=(10,8))
#plt.style.use("bmh")

g = sns.barplot(
    data = df_top10_offense_m,
    x="offense_name",
    y = "total_male",
    alpha = 0.5,
)

fig.autofmt_xdate()

# to write the number of total_male for each value 
for i in range(len(df_top10_offense_m["total_male"])):
       g.text(i, 
              i, 
              df_top10_offense_m["total_male"][i], 
              fontsize=10,
              color = "black",
              horizontalalignment='center',
             )

plt.title("Top 10 offenses by Male")
plt.xlabel("Offense Name") # x axis label
plt.ylabel("Number of Crimes") # y axis label
plt.show()
```

```python
#we group by offense_name and taking the sum of total_female , assign it to new var 
df_top10_offense_f = df_copy.groupby(["offense_name"]).total_female.sum()
# make it with descending order and take the top 10 
df_top10_offense_f = df_top10_offense_f.sort_values(ascending=False).head(10)
# here we reset the index to make it as dataframe 
df_top10_offense_f = df_top10_offense_f.reset_index()

df_top10_offense_f
```

```python
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html
# we can see a different between crimes that most common for Male and Female.

fig = plt.figure(figsize=(10,8))

g = sns.barplot(
    data = df_top10_offense_f,
    x="offense_name",
    y = "total_female",
    alpha = 0.5,
)

fig.autofmt_xdate()

# to write the number of total_male for each value 
for i in range(len(df_top10_offense_f["total_female"])):
       g.text(i, 
              i, 
              df_top10_offense_f["total_female"][i], 
              fontsize=10,
              color = "black",
              horizontalalignment='center',
             )

plt.title("Top 10 offenses by Female")
plt.xlabel("Offense Name") # x axis label
plt.ylabel("Number of Crimes") # y axis label
plt.show()
```

```python
# Here , taking three columns and all rows 
df_mean_total = df_copy.loc[:, ["year" , "total_male", "total_female"]]
# Here var_name will take the name of "total_female" and "total_male" as value
# Also, taking thier value and put them in Total column that we have created down 
df_mean_total= df_mean_total.melt("year", var_name = "Gender",value_name = "Total")

df_mean_total
```

```python
#https://seaborn.pydata.org/generated/seaborn.lineplot.html
# This the mean for Male & Female over the years 

fig = plt.figure(figsize=(10,8))

sns.lineplot(
    data=df_mean_total,
    x = "year",
    y = "Total",
   hue = "Gender"
)

plt.title("The mean crimes through the years (1994 to 2016)")
plt.xlabel("Year") # x axis label
plt.ylabel("Mean Value for Number of Crimes ") # y axis label
plt.show()
```

```python
# to take the sum for each year for male & female 
male_col = []
female_col = []

for i in range(df_mean_total["year"].min(),df_mean_total["year"].max()+1):
    male_col.append(df_mean_total[(df_mean_total["year"] == i) & (df_mean_total["Gender"] == "total_male")]["Total"].sum())
    female_col.append(df_mean_total[(df_mean_total["year"] == i) & (df_mean_total["Gender"] == "total_female")]["Total"].sum())
```

```python
# make a new dataframe that will take the three lists and assign it to nre variable 
df_total = pd.DataFrame(zip(list(range(1994, 2017)) , male_col , female_col) ,columns=['Year', 'Male' , 'Female'] )
df_total
```

```python
# the actual result for them through the years , male having a large decreased 
fig = plt.figure(figsize=(10,8))
plt.plot(df_total["Year"],df_total["Male"])
plt.plot(df_total["Year"], df_total["Female"])

plt.title("Total cimes between 1994 to 2016")
plt.xlabel("Years ") # add x axis label
plt.ylabel("Number of crimes") # add y axis label
plt.ylim(200000, 2000000)
plt.legend(["Male", "Femal"], loc ="upper right")

plt.tight_layout()
```

```python
# take the sum for each race & and make new labels with each race 
A = df_copy["white"].sum()
B = df_copy["black"].sum()
C = round(df_copy["asian_pacific_islander"].sum(),2)
D = round(df_copy["american_indian"].sum(),2)

races_total = [A,B,C,D]

race_labels = ["white", "black", "asian_pacific_islander", "american_indian"]
```

```python
#https://www.pythoncharts.com/matplotlib/pie-chart-matplotlib/

# White people having large number which is 67.8% of crimes 
fig, ax = plt.subplots(figsize=(10,8))

# create the pie plot
patche, value_name , value = ax.pie(races_total, 
        labels = race_labels,
        autopct='%1.1f%%', 
        pctdistance=0.8,
        textprops={'size': 'large'},
       )

# add circle to the middle for making the  donut chart
my_circle=plt.Circle( (0,0), 0.6, color='white')
plt.gcf().gca().add_artist(my_circle)

plt.setp(value, 
         fontweight='bold',
         color='white')

ax.axis('equal')
plt.legend(race_labels, loc="upper right")
plt.title("The proportion for each race")

plt.show()
```
